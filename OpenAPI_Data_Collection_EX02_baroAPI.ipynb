````python
from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus
import pandas as pd
import json

res_id = 'storezone'
stype = 'json'
url = 'http://apis.data.go.kr/B553077/api/open/sdsc2/baroApi'
params = f'?ServiceKey={skey}&resId={res_id}&type={stype}&' + urlencode({quote_plus('key'): '9174'}) # last '&' matters
# print(url+params)

query = url + params
request = Request(query)
request.get_method = lambda: 'GET'
response_body = urlopen(request).read()
# print(response_body)
response_dict = json.loads(response_body.decode('utf-8'))
items = response_dict['body']['items']

df = pd.DataFrame(items) # if items are dict-wise
df
````

````python
# 함수화

def baro_api(skey, res_id, encode):
  url = 'http://apis.data.go.kr/B553077/api/open/sdsc2/baroApi'
  params = f'?ServiceKey={skey}&resId={res_id}&type={stype}&' + encode

  request = Request(url+params)
  request.get_method = lambda: 'GET'
  response_body = urlopen(request).read()
  response_dict = json.loads(response_body.decode('utf-8'))
  items = response_dict['body']['items']
  df = pd.DataFrame(items)

  return df
````

````python
import requests

url = 'http://apis.data.go.kr/B553077/api/open/sdsc2/baroApi'
params = {
    'key': '9174', 
    'ServiceKey': skey,
    'resId': 'storezone',
    'type': 'json',
}

res = requests.get(url, params=params)
# print(url+'?'+urlencode(params))
pd.DataFrame(res.json()['body']['items'])
````

````python
# 함수화

def baro_api(skey, res_id):
  url = 'http://apis.data.go.kr/B553077/api/open/sdsc2/baroApi'
  params = {
    'key': '9174', 
    'ServiceKey': skey,
    'resId': 'storezone',
    'type': 'json',
  }
  
  res = requests.get(url, params=params)
  return pd.DataFrame(res.json()['body']['items'])
````

````python
def baro_api(skey, res_id, new_params):
  url = 'http://apis.data.go.kr/B553077/api/open/sdsc2/baroApi'
  params = {
    'ServiceKey': skey,
    'resId': res_id,
    'type': 'json',
  }
  params.update(new_params)
  
  res = requests.get(url, params=params)
  return pd.DataFrame(res.json()['body']['items'])
````
