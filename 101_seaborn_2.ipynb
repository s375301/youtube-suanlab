````python
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
sns.set(style="whitegrid")
````

## 범주형 데이터 (Categorical Data)

### 범주형 산점도 (Categorical scatterplots)

- ```stripplot()``` (w/ ```kind="strip"```; the default)
- ```swarmplot()``` (w/ ```kind="swarm"```)

````python
penguins = sns.load_dataset("penguins")
penguins.head()
````

````python
sns.catplot(x="species", y="body_mass_g",
            data=penguins);
````

````python
sns.catplot(x="species", y="body_mass_g",
            jitter=False, data=penguins);
````

````python
sns.catplot(x="species", y="body_mass_g",
            kind="swarm", data=penguins);
````

````python
sns.catplot(x="species", y="body_mass_g",
            hue="sex", kind="swarm",
            data=penguins);
````

````python
sns.catplot(x="sex", y="body_mass_g",
            hue="species", kind="swarm",
            data=penguins);
````

````python
# X축 순서 변경하는 법

sns.catplot(x="sex", y="body_mass_g",
            hue="species", kind="swarm",
            order=["Female", "Male"],
            data=penguins);
````

````python
sns.catplot(x="body_mass_g", y="species",
            hue="island", kind="swarm",
            data=penguins);
````

````python
sns.catplot(x="species", y="body_mass_g",
            hue="sex", col="island", kind="swarm",
            data=penguins);
````

````python
# aspect

sns.catplot(x="species", y="body_mass_g",
            hue="sex", col="island", kind="swarm", aspect=.7,
            data=penguins);
````

#### 박슨플롯 (Boxen plots)

- boxplot 보다 자세한 정보

````python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
````

````python
sns.catplot(x="cut", y="price",
            kind="boxen", data=diamonds);
````

````python
sns.catplot(x="color", y="price",
            kind="boxen", data=diamonds);
````

````python
sns.catplot(x="clarity", y="price",
            kind="boxen", data=diamonds);
````

#### 바이올린 플롯 (Violin plots)

- 커널 밀도 추정과 상자 도표 결합

````python
penguins.head()
````

````python
sns.catplot(x="species", y="body_mass_g",
            hue="sex", kind="violin",
            data=penguins);
````

````python
# bw, cut from Kernel Density Func


sns.catplot(x="species", y="body_mass_g",
            hue="sex", kind="violin",
            bw=.15, cut=0,
            data=penguins);
````

````python
# split=True

sns.catplot(x="species", y="body_mass_g",
            hue="sex", kind="violin",
            split=True, data=penguins);
````

````python
# inner="stick"/"box"

sns.catplot(x="species", y="body_mass_g",
            hue="sex", kind="violin",
            inner="stick", split=True,
            data=penguins);
````

````python
# ax=g.ax

g = sns.catplot(x="species", y="body_mass_g",
                kind="violin",
                inner=None, data=penguins);

sns.swarmplot(x="species", y="body_mass_g",
              color="k", size=3,
              data=penguins, ax=g.ax);
````

````python
sns.catplot(kind="violin", data=iris);
````

````python
sns.catplot(kind="violin", orient='h', data=iris);
````

````python
sns.catplot(x="species", y="sepal_length",
            kind="violin", data=iris);
````

````python
sns.catplot(x="sepal_length", y="species",
            kind="violin", data=iris);
````

### 범주형 추정치 도표 (Categorical estimate plots)

- ```barplot()``` (w/ ```kind="bar"```)
- ```printplot()``` (w/ ```kind="point"```)
- ```countplot()``` (w/ ```kind="count"```)

#### 막대플롯 (Bar plots)

````python
mpg = sns.load_dataset("mpg")
mpg.head()
````

````python
sns.catplot(x="origin", y="mpg",
            hue="cylinders", kind="bar",
            data=mpg);
````

````python
# palette

sns.catplot(x="origin", y="horsepower",
            hue="cylinders", kind="bar",
            palette="ch:.20", data=mpg);
````

````python
sns.catplot(x="cylinders", y="horsepower",
            kind="bar",
            palette="ch:.20", data=mpg);
````

````python
sns.catplot(x="cylinders", y="horsepower",
            kind="bar", ci=None,
            palette="ch:.20", data=mpg);
````

````python
# gray line == ci (신뢰구간)
````

````python
# edgecolor

sns.catplot(x="cylinders", y="horsepower",
            kind="bar", edgecolor=".6",
            palette="ch:.20", data=mpg);
````

#### 포인트 플롯 (Point plots)

- 축의 높이를 사용하여 추정값을 인코딩하여 점 추정값과 신뢰 구간 표시

````python
titanic = sns.load_dataset("titanic")
titanic.head()
````

````python
sns.catplot(x="who", y="survived",
            hue="class", kind="point",
            data=titanic);
````

````python
sns.catplot(x="class", y="survived",
            hue="who", kind="point",
            data=titanic);
````

````python
# palette={dict}: ggplot2::scale_color_manual

sns.catplot(x="class", y="survived",
            palette={"man": "b", 'woman': "r", "child": "g"},
            markers=["^", "o", "."],
            linestyles=["-","--",":"],
            hue="who", kind="point",
            data=titanic);
````

#### (참고)

- palette
- markers
- linestyle
  - "-": solid
  - "--": dashed
  - ":": dotted

````python
sns.catplot(x="embark_town", y="survived",
            hue="who", kind="point",
            data=titanic);
````
